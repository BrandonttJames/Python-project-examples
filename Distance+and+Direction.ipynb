```python
#the following was used to print an array of directions for a robot to travel towards a specified goal in the array

from math import *

#to hold distances from goal
a = []

for i in range(0,10):
    a.append([])
    for j in range(0,10):
        a[i].append(0)
        
goalx = 7
goaly = 7

for i in range(0,10):
    for j in range(0,10):
        a[i][j] = sqrt((goaly - 1 - i)**2 + (goalx - 1 -j)**2)

f1=open('attractive_result.txt', 'w')

for i in range(0,10):
    for j in range(0,10):
        f1.write(str(round(a[i][j], 1)))
        f1.write(" ")
    f1.write("\n")
    
f1.close()

#manhattan value array
b = []

for i in range(0,10):
    b.append([])
    for j in range(0,10):
        b[i].append(0)


for i in range(0,10):
    for j in range(0,10):
            b[i][j] = abs(goaly - 1 -i)+abs(goalx - 1 -j)

#direction array
c = []

for i in range(0,10):
    c.append([])
    for j in range(0,10):
        c[i].append(0)

t1=0
t2=0
t3=0
t4=0


for i in range(0,10):
    for j in range(0,10):
        if(i==0):
            if(j==0):
                t1=b[i+1][j]
                t2=b[i][j+1]
                if(min(t1,t2)==t1):
                    c[i][j]="down"
                else:
                    c[i][j]="right"
            elif(j==9):
                t1=b[i][j-1]
                t2=b[i+1][j]
                if(min(t1,t2)==t1):
                    c[i][j]="left"
                else:
                    c[i][j]="down"
            else:
                t1=b[i][j+1]
                t2=b[i][j-1]
                t3=b[i+1][j]
                if(min(t1,t2,t3)==t1):
                    c[i][j]="right"
                elif(min(t1,t2,t3)==t2):
                    c[i][j]="left"
                else:
                    c[i][j]="down"
        elif(j==0):
            if(i==9):
                t1=b[i][j+1]
                t2=b[i-1][j]
                if(min(t1,t2)==t1):
                    c[i][j]="right"
                else:
                    c[i][j]="up"
            else:
                t1=b[i+1][j]
                t2=b[i][j+1]
                t3=b[i-1][j]
                if(min(t1,t2,t3)==t1):
                    c[i][j]="down"
                elif(min(t1,t2,t3)==t2):
                    c[i][j]="right"
                else:
                    c[i][j]="up"
        elif(i==9):
            if(j==9):
                t1=b[i-1][j]
                t2=b[i][j-1]
                if(min(t1,t2)==t1):
                    c[i][j]="up"
                else:
                    c[i][j]="left"
            else:
                t1=b[i-1][j]
                t2=b[i][j+1]
                t3=b[i][j-1]
                if(min(t1,t2,t3)==t1):
                    c[i][j]="up"
                elif(min(t1,t2,t3)==t2):
                    c[i][j]="right"
                else:
                    c[i][j]="left"
        elif(j==9):
            t1=b[i-1][j]
            t2=b[i][j-1]
            t3=b[i+1][j]
            if(min(t1,t2,t3)==t1):
                c[i][j]="up"
            elif(min(t1,t2,t3)==t2):
                c[i][j]="left"
            else:
                c[i][j]="down"
        else:
            t1=b[i-1][j]
            t2=b[i][j-1]
            t3=b[i+1][j]
            t4=b[i][j+1]
            if(min(t1,t2,t3,t4)==t1):
                c[i][j]="up"
            elif(min(t1,t2,t3,t4)==t2):
                c[i][j]="left"
            elif(min(t1,t2,t3,t4)==t3):
                c[i][j]="down"
            else:
                c[i][j]="right"

c[goaly-1][goalx-1] = "goal"                

for i in range(0,10):
    print(c[i])
```
